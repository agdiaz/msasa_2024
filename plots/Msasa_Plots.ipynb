```python
import matplotlib.pyplot as plt

# Group the data by 'Group Length' and 'Software', then get 'Overlap Score'
grouped = df_merged.groupby(['Group Length', 'Software'])

# Define a figure to hold the subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=True)

# Names of groups and axes for easy indexing
group_names = ['CL', 'ML', 'LL']

# Loop over each group (CL, ML, LL)
for ax, group_name in zip(axes, group_names):
    # Prepare data for the boxplot: list of scores for each software in the current group
    data_to_plot = [group['Overlap Score'].values for _, group in grouped if _[0] == group_name]
    software_names = [_[1] for _ in grouped.groups if _[0] == group_name]

    # Create boxplot
    ax.boxplot(data_to_plot)
    ax.set_title(f'Grupo: {group_name}')
    ax.set_xticklabels(software_names, rotation=45, ha='right')
    ax.set_xlabel('Software')
    if ax == axes[0]:
        ax.set_ylabel('Overlap Score')

   # Identify the best performing 'MSASA' software
    msasa_medians = {}
    others_medians = {}
    for software, scores in zip(software_names, data_to_plot):
        if software.startswith("MSASA"):
            msasa_medians[software] = np.median(scores)
        else:
            others_medians[software] = np.median(scores)

    # Find the software with the highest median score
    if msasa_medians:
        best_msasa_software = max(msasa_medians, key=msasa_medians.get)

        # Highlight the best software name
        for label in ax.get_xticklabels():
            if label.get_text() == best_msasa_software:
                label.set_color('blue')  # Change color to red
                label.set_weight('bold')  # Make font bold
                label.set_bbox(dict(facecolor='yellow', edgecolor='none', alpha=0.5))
            elif label.get_text().startswith("MSASA"):
                label.set_bbox(dict(facecolor='yellow', edgecolor='none', alpha=0.5))

    # Find the software with the highest median score
    if others_medians:
        worst_others_software = min(others_medians, key=others_medians.get)
        best_others_software = max(others_medians, key=others_medians.get)

        # Highlight the best software name
        for label in ax.get_xticklabels():
            if label.get_text() == worst_others_software:
                label.set_color('red')  # Change color to red
                label.set_weight('bold')  # Make font bold
            elif label.get_text() == best_others_software:
                label.set_color('green')  # Change color to red
                label.set_weight('bold')  # Make font bold

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df_merged' is your DataFrame

# Sort the DataFrame by 'Group Length' and 'Group Quantity' for consistent ordering in plots
df_merged.sort_values(by=['Group Length', 'Group Quantity', 'Caso de Estudio'], inplace=True)

# Get unique 'Caso de Estudio' sorted by 'Group Length' and 'Group Quantity'
unique_cases = df_merged['Caso de Estudio'].unique()
cmap = plt.get_cmap('coolwarm')
norm = mcolors.Normalize(vmin=0, vmax=1)

# Create a plot for each 'Caso de Estudio'
for study_case in unique_cases:
    # Filter data for the current case
    data = df_merged[df_merged['Caso de Estudio'] == study_case]

    # Prepare data for plotting
    software_names = data['Software']
    scores = data['Overlap Score']

    sorted_indices = scores.argsort()[::-1]  # Sort scores from high to low
    sorted_software_names = software_names.iloc[sorted_indices]
    sorted_scores = scores.iloc[sorted_indices]

    # Create horizontal bar plot
    plt.figure(figsize=(10, 6))
    bars = plt.barh(sorted_software_names, sorted_scores, color=[cmap(norm(score)) for score in sorted_scores])
    plt.xlabel('Overlap Score')
    plt.xlim(0, 1)  # Set x-axis limits to 0 to 1
    plt.title(f'Caso de estudio {data.iloc[0]["Study Case"]} - Secuencia: {study_case} - Longitud máxima: {data.iloc[0]["Dimension Length"]} - '
              f'Cantidad de secuencias: {data.iloc[0]["Dimension Quantity"]} - '
              f'Grupos: {data.iloc[0]["Group Length"]} - {data.iloc[0]["Group Quantity"]}')

    # Annotate bars with their values
    for bar, value in zip(bars, sorted_scores):
        # Determine text position inside or outside the bar based on value
        text_x = bar.get_width() if bar.get_width() < 0.95 else bar.get_width() - 0.05
        text_color = 'black' if bar.get_width() < 0.95 else 'white'
        plt.text(text_x, bar.get_y() + bar.get_height()/2, f' {value:.2f}',
                 va='center', ha='left', color=text_color)

    for bar, software_name in zip(bars, sorted_software_names):
        if software_name.startswith("MSASA"):
            plt.gca().get_yticklabels()[bars.index(bar)].set_bbox(dict(facecolor='yellow', edgecolor='none', alpha=0.5))

    plt.gca().invert_yaxis()  # Invert y-axis to show highest scores at the top
    plt.show()
    # break
```
